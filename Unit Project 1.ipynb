## Problem Statement

Determine the likelihood of an applicant being admitted to graduate school at UCLA, using the applicant's GRE score, GPA, and the prestige of the applicant's undergraduate institution. The data comes from UCLA's Logit Regression in R tutorial.

Hypothesis: An applicant's GRE score, GPA, and the prestige of the applicant's undergraduate institution are all positively correlated with admission to graduate school at UCLA.

## Risks and assumptions

The data includes admissions decisions for 400 applicants to graduate school at UCLA. UCLA admits an average of 6000 students out of 21,300 who apply each year (see http://www.ucla.edu/admission/graduate-admission). Consequently, the data represents less than 2% of the applicant pool in a typical year. For purposes of this project, we are assuming that this is a representative sample of UCLA's graduate student applicant pool.

However, the admissions data is not broken down by department, and it is unknown whether graduate student admissions rates and/or standards vary significantly by academic department at UCLA. Furthermore, it is not clear if this represents applicants in a single or multiple years. If applicant data comes from multiple years, it is possible that changing trends in admission patterns may affect the results. 

Applicants also submit personal statements, recommendations, portfolios, and other documents when applying to graduate school. The data does not capture the effect, if any, of these other materials upon admissions decisions at UCLA.

Although GPA is recorded using the familiar 4-point scale for American schools, grading standards vary from school to school. It does not appear that GPA scores have been adjusted to account for inconsistency in grading standards among different schools. 

It is unknown how UCLA determined the prestige of each applicant's undergraduate institution. For purposes of this project, we are taking it on faith that UCLA's admissions staff correctly determined the prestige of each applicant's undergraduate institution.
