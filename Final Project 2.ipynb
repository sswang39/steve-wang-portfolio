# Final Project 2

## Problem

The aim is to construct a machine learning model that will predict the sale price of residential homes, using features of each property (e.g., square footage, year of construction, etc.). The model would predict a continuous number (i.e., the sale price), and so the problem is  a regression problem.

This would help both buyers and sellers of residential homes. Sellers would have a more accurate expectation of how much they might receive from the sale of a home. Furthermore, I hope the model will identify the features that have the greatest impact upon the sale price of a home. Understanding which features most influence the sale price of a home would assist sellers in marketing their homes. Furthermore, buyers would also benefit by gaining a better expectation of what kind of home they might realistically be able to purchase for a given amount. 

## Hypothesis

The sale price of a home increases (1) as the square footage of the total living area increases, and (2) the more recently it was built.

## Dataset

The project uses the Ames Housing dataset, which is available at Kaggle (https://www.kaggle.com/c/house-prices-advanced-regression-techniques).

The Ames Housing dataset describes 79 features of 2920 residential homes sold in Ames, Iowa, from 2006 through 2010. (Kaggle splits the dataset into training and test sets of equal size, i.e., 1460 observations each.)

A description of the features is available at https://kaggle2.blob.core.windows.net/competitions-data/kaggle/5407/data_description.txt?sv=2015-12-11&sr=b&sig=jee8dIVA6iOAXvBaINaElc38GDUzp%2FGWgp5%2B8PWZXe8%3D&se=2017-01-11T20%3A46%3A56Z&sp=r

## Project Concerns

My first concern is getting a good grasp of the dataset and its features. I have started exploring the dataset and am gradually getting a feel for it. But there's a lot of data to wrap my head around.

The large number of features makes the curse of dimensionality a serious concern. I've also observed that the dataset contains a lot of categorical values that will have to be converted to dummy variables. This may result in a sparse matrix--I'm already seeing numerous feature values that are only present in a small number of houses in the training set.

Linear regression is the obvious starting point. Lasso will be needed to identify the most important features.

## Outcomes

The desired outcome is a machine learning model that predicts the sale prices of homes in the test set with a low mean average error. 

Ideally, this would be a model that is based on a manageably small number of features--i.e., the features that have the greatest influence upon the sale price of a home. I feel that identifying these features would provide the most benefit by illuminating the most important determinants of a home's sale price.
